```python
#Question 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.
import numpy as np
import matplotlib.pyplot as plt
mean_target = 50
std_dev_target = 5
samples = 1000
data = np.random.normal(mean_target, std_dev_target, samples)
calculated_mean = np.mean(data)
calculated_std_dev = np.std(data)
print(f"Calculated Mean: {calculated_mean:.2f}")
print(f"Calculated Standard Deviation: {calculated_std_dev:.2f}")
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
```

```python
#Question 10: You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.● Write the Python code to compute the mean sales and its confidence interval.

import numpy as np
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250]
mean = np.mean(daily_sales)
n = len(daily_sales)
std_dev = np.std(daily_sales, ddof=1)
se = std_dev / np.sqrt(n)
z_critical = 1.96
lower_bound = mean - (z_critical * se)
upper_bound = mean + (z_critical * se)
print(f"Sample Mean: {mean:.2f}")
print(f"Standard Error: {se:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```
